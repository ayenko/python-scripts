```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
file_path_1 = '/content/piano_clean_up_all_uers.csv' #All Users PIANO - 133k
file_path_2 = '/content/invalid emails.csv' #Invalid List From SailThru - 25000
file_path_3 = '/content/MIXPANEL PIANO DATA.csv' #From MixPanel - 133k
```

```python
# Read the CSV file into a pandas DataFrame
df_a = pd.read_csv(file_path_1, encoding='utf-8')
df_b = pd.read_csv(file_path_2, encoding='utf-8')
df_c = pd.read_csv(file_path_3, encoding='utf-8')
```

```python
# Merge df_a with df_c to get 'Membership' and 'Last seen' columns
df_a = df_a.merge(df_c[['User ID', 'membership', 'Last seen']], on='User ID', how='left')

# Convert 'Since' and 'Last seen' columns to datetime objects
df_a['Since'] = pd.to_datetime(df_a['Since'], format='%d/%m/%Y %H:%M')
df_a['Last seen'] = pd.to_datetime(df_a['Last seen'], format='%d/%m/%Y %H:%M')

# Calculate 'Last 12 Months' column
twelve_months_ago = datetime.now() - timedelta(days=365)
df_a['Last 12 Months'] = df_a['Last seen'].apply(lambda x: 'Yes' if x >= twelve_months_ago else 'No')

# Merge df_a with df_b to get 'Email Status' column
df_a['Email Status'] = df_a['User ID'].isin(df_b['User ID']).apply(lambda x: 'Invalid' if x else 'Valid')

# Export updated df_a as a CSV
df_a.to_csv('updated_df_a.csv', index=False)
```
