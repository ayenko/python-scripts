```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import seaborn as sns
import numpy as np
```

```python
file_path = '/content/June Reach Revenue Reputation Analysis.csv'
```

```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, encoding='utf-8')
```

```python
df.head()
```

```python
# Create a function to determine the Article type based on the combination of Reach, Revenue, and Reputation.
def determine_article_type(row):
    if row['Reach'] == 1 and pd.isna(row['Revenue']) and pd.isna(row['Reputation']):
        return 'Reach'
    elif pd.isna(row['Reach']) and row['Revenue'] == 1 and pd.isna(row['Reputation']):
        return 'Revenue'
    elif pd.isna(row['Reach']) and pd.isna(row['Revenue']) and row['Reputation'] == 1:
        return 'Reputation'
    elif row['Reach'] == 1 and row['Revenue'] == 1 and pd.isna(row['Reputation']):
        return 'Reach + Revenue'
    elif row['Reach'] == 1 and pd.isna(row['Revenue']) and row['Reputation'] == 1:
        return 'Reach + Reputation'
    elif pd.isna(row['Reach'])  and pd.isna(row['Revenue']) and row['Reputation'] == 1:
        return 'Revenue + Reputation'
    else:
        return 'Other'

# Apply the function to create a new column "Article type."
df['Article type'] = df.apply(determine_article_type, axis=1)

# Group the DataFrame by "Article type" and calculate counts, total Pageviews, and average Pageviews.
result_df = df.groupby('Article type').agg(
    Count=('Pageviews', 'count'),
    Total_Pageviews=('Pageviews', 'sum'),
    Avg_Pageviews=('Pageviews', 'mean'),
    U_Pageviews=('U Pageviews', 'mean')
).reset_index()

# Rename the 'Pageviews' columns to 'Avg Pageviews.'
result_df.rename(columns={'Count': 'Article Type Count'}, inplace=True)

# Display the resulting DataFrame.
print(result_df)
```

```python
# Export the DataFrame to a CSV file.
result_df.to_csv("article_type_summary.csv", index=False)
```

```python
# Create columns for Org Type 1 and Org Type 2 based on specified combinations. - Service provider to Startups
df['Org Type 1'] = df['Corporate'] + df['Investment firm'] + df['Service provider'] + df['Series B+ Startup']
df['Org Type 2'] = df['Pre-seed Startup'] + df['Seed Startup'] + df['Series A Startup']
df['Other Org Types'] = df['I dont work at the moment'] + df['Other org types']
df['Seniority Type 1'] = df['Director'] + df['Senior manager'] + df['Partner']
df['Seniority Type 2'] = df['(Co) Founder']
df['Other Seniority Types'] = df['Associate'] + df['Manager'] + df['Student'] + df['Academic'] + df['Other roles'] + df['undefined']+ df['Junior']

# Group the DataFrame by "Article type" and calculate the average Pageviews, Org Type 1, and Org Type 2.
result_df = df.groupby('Grouped topic').agg(
    Article_Count=('Pageviews', 'count'),
    Total_Pageviews=('Pageviews', 'sum'),
    Avg_Pageviews=('Pageviews', 'mean'),
    U_Pageviews=('U Pageviews', 'mean'),
    Org_Type_1=('Org Type 1', 'mean'),
    Org_Type_2=('Org Type 2', 'mean'),
    Other_Org_Types=('Other Org Types', 'mean'),
    Seniority_Type_1=('Seniority Type 1', 'mean'),
    Seniority_Type_2=('Seniority Type 2', 'mean'),
    Other_Seniority_Types=('Other Seniority Types', 'mean'),
).reset_index()

# Rename the 'Pageviews' columns to 'Avg Pageviews.'
result_df.rename(columns={'Article type': 'Article Types'}, inplace=True)

# Display the resulting DataFrame.
print(result_df)
```

```python
# Export the DataFrame to a CSV file.
result_df.to_csv("may_article_type_summary.csv", index=False)
```
