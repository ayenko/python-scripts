```python
import pandas as pd
```

```python
file_path = 'startup-life-newsletter-data.csv'
nl_data = pd.read_csv(file_path)
```

```python
nl_data.head()
```

```python
nl_data['merged_country'] = nl_data.apply(lambda row: row['Country'] if pd.isnull(row['country_code']) else row['country_code'], axis=1)
```

```python
nl_data.head()
```

```python
file_path = 'all_users.csv'
users_data = pd.read_csv(file_path)
```

```python
nl_data.shape[0]
```

```python
users_data.shape[0]
```

```python
users_data.head()
```

```python
merged_df = pd.merge(nl_data, users_data, how='left', left_on='Email', right_on='email')
```

```python
merged_df.shape[0]
```

```python
merged_df.head()
```

```python
# Fill missing values in the 'merged_country' column using data from 'country_code' in 'users_data'
merged_df['merged_country'] = merged_df.apply(lambda row: row['country_code_y'] if pd.isnull(row['merged_country']) else row['merged_country'], axis=1)
```

```python
merged_df.head()
```

```python
# Export the merged DataFrame to a CSV file
merged_df.to_csv('merged_data.csv', index=False)
```
