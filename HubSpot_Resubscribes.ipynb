```python
#import pandas library
import pandas as pd
import re
```

```python
#change this to the name of the exported csv file with the necceary columns (pianoid, error)
file_path = '/content/resubscribes.csv'
```

```python
# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, encoding='utf-8')
```

```python
df.head()
```

```python
# Extract digits after VID= and create 'Hubspot ID' column
df['Hubspot ID'] = df['error'].apply(lambda x: re.search(r'VID=(\d+)', x).group(1) if re.search(r'VID=(\d+)', x) else None)
```

```python
# Function to check if a string is present in the error column
def check_list(error, keyword):
    if keyword in error:
        return "TRUE"
    else:
        return ""
```

```python
# Add new columns based on the logic provided
df["Sifted Marketing Updates"] = df["error"].apply(lambda x: check_list(x, "would-like-event-newsletter"))
df["The Sifted Newsletter"] = df["error"].apply(lambda x: check_list(x, "wants_main_newsletter"))
df["Startup Life Newsletter"] = df["error"].apply(lambda x: check_list(x, "would-like-startup-life-newsletter"))
df["Fintech Newsletter"] = df["error"].apply(lambda x: check_list(x, "would-like-fintech-newsletter"))
df["Climate Tech Newsletter"] = df["error"].apply(lambda x: check_list(x, "would-like-sustainability-newsletter"))
df["Up Round Newsletter"] = df["error"].apply(lambda x: check_list(x, "would-like-futureproof-newsletter"))
df["Event sponsorship opportunities"] = df["error"].apply(lambda x: check_list(x, "Events sponsorship interest"))
df["Sifted Events"] = df["error"].apply(lambda x: check_list(x, "Sifted Events"))
```

```python
# Merge rows with the same 'pianoId'
#df = df.groupby('pianoId').agg(lambda x: ''.join(x)).reset_index()

# Consolidate TRUE values for each row
df = df.groupby('pianoId').agg({
    'Hubspot ID': 'first',
    'Sifted Marketing Updates': 'max',
    'The Sifted Newsletter': 'max',
    'Startup Life Newsletter': 'max',
    'Fintech Newsletter': 'max',
    'Climate Tech Newsletter': 'max',
    'Up Round Newsletter': 'max',
    'Event sponsorship opportunities': 'max',
    'Sifted Events': 'max'
}).reset_index()
```

```python
# Save the updated DataFrame to a new CSV file
df.to_csv("HubSpot List - updated.csv", index=False)
```
