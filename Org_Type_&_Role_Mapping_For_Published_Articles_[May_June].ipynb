```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import seaborn as sns
import numpy as np
```

```python
file_path1 = '/content/june_article_seg.csv'
file_path2 = '/content/June Articles Role.csv'
```

```python
# Read the CSV 1 file into a pandas DataFrame
df1 = pd.read_csv(file_path1, encoding='utf-8')
df2 = pd.read_csv(file_path2, encoding='utf-8')
```

```python
df1.head()
```

```python
df2.head()
```

```python
# Merge the dataframes on 'Page path' to combine the data
merged_df = df1.merge(df2, on='Page path', how='left')
```

```python
merged_df.head()
```

```python
# Calculate page views split by Organization type for each unique page path
unique_page_paths = merged_df['Page path'].unique()
results = []

for path in unique_page_paths:
    page_views_by_type = merged_df[merged_df['Page path'] == path].groupby('role')['Page view'].sum().reset_index()
    results.append({'Page path': path, **dict(page_views_by_type.values)})

# Create a new dataframe with the results
final_result = pd.DataFrame(results)
```

```python
final_result.head()
```

```python
# Export the results to a CSV file
final_result.to_csv('june_role.csv', index=False)
```
